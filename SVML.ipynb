```python
X=df.drop(['target','fname'],axis = 'columns')
```

```python
y = df.target
```

```python
y
```

```python
#Splitting data into train and test
```

```python
# importinf train test split from sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
#total rows = 150
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_test
```

```python
#imorting support vector machine learning
```

```python
from sklearn.svm import SVC 
```

```python
# here SVC --> support vector classifier 
```

```python
svmodel = SVC(kernel = 'linear')
```

```python
svmodel.fit(X_train,y_train)
```

```python
svmodel.predict([[6.7,3.3,5.7,2.1]])
```

```python
svmodel.predict([[3.5,7.1,0.5,7.7]])
```

```python
y_predict = svmodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_predict,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_predict,y_test)
```
